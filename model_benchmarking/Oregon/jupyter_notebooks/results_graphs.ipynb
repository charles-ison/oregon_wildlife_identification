```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = ("All Images", "Sufficient Lighting", "Insufficient Lighting", "Winter", "Batching")
model_accuracy = {
    'ResNet-50': (97.60, 99.21, 84.26, 90.91, 98.11),
    'ResNet-152': (97.30, 98.32, 88.88, 91.74, 99.06),
    'ViT-L-16': (97.10, 98.06, 88.89, 90.01, 0),
}

x = np.arange(len(categories))
width = 0.25
multiplier = 0

fig, ax = plt.subplots(figsize=(10, 7), constrained_layout=True)

for model, accuracy in model_accuracy.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, accuracy, width, label=model)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_ylabel('Accuracy (%)')
ax.set_title('Model Performance by Image Type')
ax.set_xticks(x + width, categories)
ax.legend(loc='upper center', ncols=3)
ax.set_ylim(80, 100)

plt.show()
```
